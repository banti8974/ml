```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("C:\\Users\\nitis\\Downloads\\DigitalDesignCSV.csv")
```

```python
class_counts = df['Final Grade'].value_counts()
```

```python
plt.figure(figsize=(8, 6))
class_counts.plot(kind='bar', color='skyblue')
plt.title('Distribution of Final Grades')
plt.xlabel('Final Grade')
plt.ylabel('Frequency')
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
imbalance_threshold = 0.1  # Adjust threshold as needed
imbalanced = any((count / len(df)) < imbalance_threshold for count in class_counts)
if imbalanced:
    print("Class imbalance detected in the 'Final Grade' column.")
else:
    print("No class imbalance detected in the 'Final Grade' column.")
```
