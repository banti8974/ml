```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder
```

```python
df = pd.read_csv("C:\\Users\\nitis\\Downloads\\DigitalDesignCSV.csv")
```

```python
label_encoder = LabelEncoder()
```

```python
df['Final Grade_encoded'] = label_encoder.fit_transform(df['Final Grade'])
```

```python
print("Mapping of original labels to encoded numerical values:")
for original_label, encoded_value in zip(label_encoder.classes_, label_encoder.transform(label_encoder.classes_)):
    print(f"{original_label} -> {encoded_value}")
```
